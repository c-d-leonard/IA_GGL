```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rp = np.loadtxt('./txtfiles/boosts/rpvec.txt')
B_a = np.loadtxt('./txtfiles/boosts/Boost_A_survey=DESY1_DESHoD.txt')
B_b = np.loadtxt('./txtfiles/boosts/Boost_B_survey=DESY1_DESHoD.txt')
```

```python
plt.figure()
plt.loglog(rp, B_a,'o', label='bin0')
plt.loglog(rp, B_b,'o', label='bin1')
plt.ylabel('B-1')
plt.xlabel('$r_p$, Mpc/h')
plt.legend()
#plt.savefig('./txtfiles/boosts/boosts_DESY1.png')
plt.show()
```
